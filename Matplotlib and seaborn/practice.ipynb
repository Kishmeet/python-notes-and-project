```python
import matplotlib.pyplot as plt
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
kohli =  [0,0, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
rohit= [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
plt.plot(years,kohli,color="red",linestyle='--',label="kohli")
plt.plot(years,rohit,color="blue",linestyle='-.',label="rohit")
plt.plot(years,sehwag,color="green",linestyle=':',label="sehwag")
plt.xlabel("years")
plt.ylabel("runs")
plt.legend()
```

```python
import numpy as np
years=[2008,2009,2010,2011,2012]
sachin = [1200, 1700, 1300, 900, 1500]
sehwag = [ 1800, 1500, 1100, 800, 0]
kohli  = [ 1300, 1500, 1800, 1900, 2100]

x = np.arange(len(years))  # index positions
width = 0.25
plt.bar(x-width,kohli,width=width,label='Kohli')
plt.bar(x,sehwag,width=width,label='sehwag')
plt.bar(x+width,sachin,width=width,label='sachin')
plt.xlabel("years")
plt.ylabel("runs")
plt.xticks(x,years)
plt.legend()
plt.xticks(x,years)
```

```python
players=["Sachin", "Sehwag", "Kohli"]
runs=[500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100]
plt.barh(players,runs)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid") 
import numpy as np
x = np.array([0, 2, 3, 4, 5, 6, 7, 8, 9 ,10])
y = np.sin(x)
sns.lineplot(x=x, y=y)
plt.title('Beautiful Line Plot')
plt.show()
```

```python
flights = sns.load_dataset('flights')
pivot_table = flights.pivot(index="month", columns="year", values="passengers")
sns.heatmap(pivot_table, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Passengers')
```
